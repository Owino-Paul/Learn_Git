```python
import pandas as pd
products = ['Apple', 'Banana', 'Carrot', 'Daikon', 'Avocado', 'Orange']
product_series = pd.Series(products)
print(product_series)
```

```python
price = [50, 20, 30, 100, 25, 50,]
price_series = pd.Series(price)
print(price_series)
```

```python
# Join the two series together
product_price = pd.concat([product_series, price_series], axis=1)
print(product_price)
```

```python
# combine the series
combine_series = product_series + " KeS " + price_series.astype(str)
print(combine_series)
```

```python
# Create a DataFrame from a list od dictionary
data = {'Product': product_series, 'Price': price_series}
df1 = pd.DataFrame(data)
print(df1)
```

```python
# Create a DataFrame from a list of dictionary
data = [{'Product': products[i],'Price': price [i]} for i in range (len(products))]
df2 = pd.DataFrame(data)
print(df2)
```
