### 8. Transform the 'counter_type' feature to a numeric variable using the encoder of your choice.

```python
# transform column ['counter_type'] to a numeric variable
# client_bills = pd.get_dummies(client_0_bills, columns=['counter_type'], prefix='counter')
# print(client_bills)

print(client_0_bills.columns)

```

```python
# transform column ['counter_type'] to a numeric variable using Label Encoding
from sklearn.preprocessing import LabelEncoder

# Instantiate the LabelEncoder
label_encoder = LabelEncoder()

# Assuming 'counter_type' column exists, fit_transform and create a new column
# If 'counter_type' is dropped, you'll need to reload the data or recreate the column
if 'counter_type' in client_0_bills.columns:
    client_0_bills['counter_type_encoded'] = label_encoder.fit_transform(client_0_bills['counter_type'])
else:
    print("Column 'counter_type' not found. It may have been dropped previously.")

# Print the updated columns of the DataFrame
print(client_0_bills.columns)
```
