```python
# Introduction to File Handling In Python
# 1. Opening a file
# 2. Closing a file

file = open("practice.txt", "w")
content = file.write("Hello World")
print(content)

file.close()
```

```python
# Reading the file we created
file = open("practice.txt", "r")
content = file.read()
print(content)

file.close()
```

```python
# Using the WITH STATEMENT
# Here we don't need the .close statement.
# The file closes immediately(automatically).

with open("practice.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# Create the same file
file = open("practice.txt", "x")
content = file.create("Hello World")
print(content)

file.close()
```

```python
# Overwrite/change the content of file
file = open("practice.txt", "w")
content = file.write("Owino Paul")
print(content)

file.close()
```

```python
file = open("practice.txt", "r")
content = file.read()
print(content)

file.close()
```

```python
with open("practice.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
# Appending or adding something to the existing file
file = open("practice.txt", "a")
content = file.write("\nDay Walker")
print(content)

file.close()
```

```python
# Slicing the file
# This method is the most efficient even for large files
with open("practice.txt", "r") as file:
    print(file.read(6))
```

```python
# Slicing the file
# But this method is inefficient especially for large files
with open("practice.txt", "r") as file:
    print(file.read()[0:6])
```

```python
# Writng a  CSV Files
import csv
with open("practice.csv", "w", newline = '') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Age', 'Country'])
    writer.writerow(['John Doe', '25', 'Kenya'])
    writer.writerow(['Jane Doe', '30', 'USA'])
    writer.writerow(['Bob Smith', '40', 'Canada'])
    writer.writerow(['Alice Johnson', '28', 'Australia'])
    writer.writerow(['Tom Wilson', '35', 'Germany'])
    writer.writerow(['Owino Paul', '59', 'Kenya'])
```

```python
# Reading the CSV file
with open("practice.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
