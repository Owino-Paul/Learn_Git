```python
# Import Pandas
import pandas as pd

# Read a sample CVS file
df = pd.read_csv('Loan_prediction_dataset (1).csv')

#check first 5 records
df.head(5)
```

```python
# checking last five details
df.tail(5)
```

```python
# checking the details of the data
df.info()
```

```python
# Describing the data set
df.describe()

```

```python
# Count NUll values in the data
df.isnull().sum()
```

```python
# Filter through our data
df.loc[df['Gender'] == 'Male']
```

```python
# Filtering for multiple results
df.loc[(df['Gender'] == 'Male') & (df['Education'] == 'Graduate')]
```

```python
# Filter using Querry
df.query('Gender == "Male"')
```

```python
# Using filtered to view mutliple fields.
df_filtered = df.query('Gender == "Female" and Education == "Graduate" and Loan_Status == "Y"')
df_filtered
```

```python
# Shaping the dataset
df_filtered.shape
```

```python
# checking if there are null values
df.isnull().sum()
```

```python
# Getting Mean of ApplicantIncome
df['ApplicantIncome'].mean()
```

```python
# Mean for CoapplicantIncome
df['CoapplicantIncome'].mean()
```

```python
# # Using filtered to view mutliple fields - Married, Self_Employed, Property_Area
df_filtered = df.query('Married == "Yes" and Self_Employed == "No" and LoanAmount >= 250 and Property_Area == "Rural"')
df_filtered
```

```python
# Importing MatPlotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Simple Visualisation
sns.countplot(x='Gender', data=df)
plt.show()
```

```python
# visulisation in bar graph
sns.barplot(x='Gender', y=df['Gender'].value_counts(), data=df)
plt.show()


```

```python
# visulisation pie chart
df['Gender'].value_counts().plot.pie()
plt.show()
```

```python
# groupby gender
df.groupby('Gender')['ApplicantIncome', CoapplicantIncome].mean() .plot.bar(kind = 'bar').set_title()
plt.show()
```
